```python
udpxy_prefix='http://192.168.10.250:8188/udp/'
with open('江苏电信-组播-全部_zsd.m3u', 'w', encoding='utf-8') as file:
    file.write(f"#EXTM3U\n")
    for index, row in df.iterrows():
        file.write(f"#EXTINF:-1 tvg-id=\"{index+1}\" tvg-name=\"{row['频道名称']}\" group-title=\"{row['类型']}\",{row['频道名称']}-{row['清晰度']}\n")
        file.write(f"{row['组播地址'].replace('igmp://', udpxy_prefix)}\n")
```
